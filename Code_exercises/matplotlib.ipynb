```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20 ,5)
y = 2 * x +1
```

```python
plt.scatter(x,y, s= 200, marker = 's')
plt.show()
```

```python
plt.xlim(0,30)
plt.ylim(0,30)
x_ticks = np.linspace(0,30,11)
plt.xticks(x_ticks)

plt.scatter(x,y, s= 200, marker = 's')
plt.show()
```
